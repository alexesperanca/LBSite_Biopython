```python
from Bio import Entrez
from Bio import SeqIO

#Aceder ao NCBI
Entrez.email = 'pg45468@alunos.uminho.pt'
handle = Entrez.efetch(db = "nucleotide", id = "AL123456.3", rettype = "gb", retmode = "text")
seq_record = SeqIO.read(handle, "gb")

#Criar um ficheiro
SeqIO.write(seq_record, "mycobacterium_tuberculosis.gb", "genbank")
handle.close()

#Aceder aos dados
record = SeqIO.read("mycobacterium_tuberculosis.gb", "genbank") 
print (record)
```
